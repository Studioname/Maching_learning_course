Data in scikit-learn is in most cases saved as 2d numpy arrays with the shape (n, m). Many algorithms also accept scipy.sparse matrices of the same shape.

1. n: (n_samples) The number of samples. Each sample is an item to process. A sample can be a document, picture, sound, video, astronomical object, row in database in csv, or whatever you can describe with a fixed set of quantitative traits.

2. m: (n_features) The number of features or distinct traits that can be used to describe each item in a quantitative manner. Features are generally real-valued, but may be boolean or discrete-valued in some cases.

## Visualising the Features of the Iris Data Set

The feature data is four dimensional, but we can visualise one or two of the dimensions at a time using a simple histogram or scatter-plot.

```python
iris.data[iris.target==1][:5]
```

```python
import matplotlib.pyplot as plt
```

```python
# let's make a histogram for our data set

fig,ax = plt.subplots()
x_index = 3
colors = ['blue', 'red', 'green']

for label, color in zip(range(len(iris.target_names)), colors):
    ax.hist(iris.data[iris.target==label, x_index],
           label=iris.target_names[label],
           color=color)
    
ax.set_xlabel(iris.feature_names[x_index])
ax.legend(loc='upper right')
fig.show()
```

We'll now look at all feature combinations in a single combined diagram. This will automatically create histograms of all possible feature pairings for our data.

To do this we will pull the length of the dataset's feature names, and simply perform a nested loop through it.

```python
n = len(iris.feature_names)
fig, ax = plt.subplots(n, n, figsize=(16,16))

colors = ['blue', 'green', 'red']
for x in range(n):
    for y in range(n):
        xname = iris.feature_names[x]
        yname = iris.feature_names[y]
        for color_ind in range(len(iris.target_names)):
            ax[x,y].scatter(iris.data[iris.target==color_ind, x],
                           iris.data[iris.target==color_ind, y],
                           label=iris.target_names[color_ind],
                            c=colors[color_ind])
            ax[x,y].set_xlabel(xname)
            ax[x,y].set_ylabel(yname)
            ax[x,y].legend(loc='upper left')
plt.show()
```

## Loading Digits Dataset

Now we'll take a look at a dataset where we have to put a bit more thought in how to represent the data. We can explore the data in a manner similar to the our approach above.

```python
from sklearn.datasets import load_digits

digits = load_digits()
digits.keys()
```

```python
n_samples, n_features = digits.data.shape
```

```python
n_samples, n_features
```

```python
digits.data[0]
```

```python
digits.target
```

The target here is the digit represented by the data. The data is an array of length 64, but what does this data mean?

There is a clue in the fact that we have two versions of the data array: data and images. Let's take a look

```python
digits.data.shape
```

```python
digits.images.shape
```

We can see that they're related by a simple reshaping.

```python
import numpy as np
np.all(digits.images.reshape(1797, 64) == digits.data)
```

We'll now visualise the data - It's a little more involved than the scatter-plot from earlier.

```python
fig = plt.figure(figsize=(6, 6)) # figure size in inches
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05,wspace=0.05)

for i in range(64):
    ax = fig.add_subplot(8,8,i+1,xticks=[],yticks=[])
    ax.imshow(digits.images[i], cmap=plt.cm.binary, interpolation='nearest')
    ax.text(0,7, str(digits.target[i]))
```

We see now what the features mean. Each feature is a real valued quantity representing the darkness of a pixel in a 8x8 image of a hand written digit.

Even though each sample has data that is inherently two-dimensional, the datamatrix flattens this 2D data into a single vector, which can be contained in one row of the data matrix.
